```python
import collections
from datasets import load_dataset
from transformers import AutoTokenizer

# Load full WikiText-2 dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1")
train_text = dataset['train']['text']

# Tokenize using basic whitespace + frequency thresholding
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Step 1: Tokenize all words (not subword/BPE) for frequency counting
all_tokens = []
for line in train_text:
    tokens = line.strip().split()  # simple whitespace tokenization
    all_tokens.extend(tokens)

# Step 2: Count frequencies
freq_counter = collections.Counter(all_tokens)

# Step 3: Build vocab (top 10,000 words with frequency >= 5)
special_tokens = ["<PAD>", "<UNK>", "<BOS>", "<EOS>"]
min_freq = 5
max_vocab_size = 10000 - len(special_tokens)

vocab = [word for word, freq in freq_counter.items() if freq >= min_freq]
vocab = sorted(vocab, key=lambda x: -freq_counter[x])[:max_vocab_size]
final_vocab = special_tokens + vocab
word2idx = {word: idx for idx, word in enumerate(final_vocab)}

# Vocabulary stats
vocab_stats = {
    "Total unique tokens": len(freq_counter),
    "Vocabulary size (after filtering)": len(final_vocab),
    "UNK token index": word2idx["<UNK>"],
    "Top 5 frequent tokens": vocab[:5]
}

vocab_stats
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

# Step 1: Replace rare tokens with <UNK> and map to indices
token_indices = []
for line in train_text:
    tokens = line.strip().split()
    for token in tokens:
        token = token.strip()
        if token in word2idx:
            token_indices.append(word2idx[token])
        else:
            token_indices.append(word2idx["<UNK>"])

print(f"Total usable tokens: {len(token_indices)}")

# Step 2: Create Dataset class
SEQ_LEN = 8

class NextTokenDataset(Dataset):
    def __init__(self, token_ids, seq_len):
        self.inputs = []
        self.targets = []
        for i in range(len(token_ids) - seq_len):
            self.inputs.append(token_ids[i:i+seq_len])
            self.targets.append(token_ids[i+seq_len])

    def __len__(self):
        return len(self.inputs)

    def __getitem__(self, idx):
        return torch.tensor(self.inputs[idx], dtype=torch.long), torch.tensor(self.targets[idx], dtype=torch.long)

# Step 3: Instantiate dataset and dataloader
dataset = NextTokenDataset(token_indices, SEQ_LEN)
dataloader = DataLoader(dataset, batch_size=64, shuffle=True)

# Quick check
batch = next(iter(dataloader))
print("Input shape:", batch[0].shape)  # (B, SEQ_LEN)
print("Target shape:", batch[1].shape)  # (B,)
```

```python
import torch.nn as nn
import torch.optim as optim
import time

VOCAB_SIZE = len(final_vocab)  # 10,000
EMBED_DIM = 128
HIDDEN_DIMS = [256, 128]

class FFNModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dims, seq_len):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.fc1 = nn.Linear(embed_dim * seq_len, hidden_dims[0])
        self.fc2 = nn.Linear(hidden_dims[0], hidden_dims[1])
        self.output = nn.Linear(hidden_dims[1], vocab_size)
        self.relu = nn.ReLU()

    def forward(self, x):
        x = self.embedding(x)              # [B, S, E]
        x = x.view(x.size(0), -1)          # flatten: [B, S*E]
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        return self.output(x)

# Instantiate model
model = FFNModel(VOCAB_SIZE, EMBED_DIM, HIDDEN_DIMS, SEQ_LEN)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
from tqdm import tqdm

def train(model, dataloader, optimizer, criterion, epochs=3):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        loop = tqdm(dataloader, desc=f"Epoch {epoch+1}")
        for X, y in loop:
            # ✅ Move to GPU
            X, y = X.to(DEVICE), y.to(DEVICE)

            # ✅ Sanity check (optional)
            assert next(model.parameters()).is_cuda, "Model is not on GPU"
            assert X.is_cuda and y.is_cuda, "Data not on GPU"

            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            loop.set_postfix(batch_loss=loss.item())

        avg_loss = total_loss / len(dataloader)
        print(f"[Epoch {epoch+1}] Avg Loss: {avg_loss:.4f}")
```

```python
train(model, dataloader, optimizer, criterion, epochs=3)
```

```python
def evaluate(model, dataloader):
    model.eval()
    correct1 = correct3 = correct5 = total = 0
    start_time = time.time()

    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(DEVICE), y.to(DEVICE)
            outputs = model(X)  # [B, vocab_size]
            _, top5 = outputs.topk(5, dim=1)  # top-5 predicted token indices
            correct = top5.eq(y.view(-1, 1))  # compare to ground truth

            correct1 += correct[:, 0].sum().item()
            correct3 += correct[:, :3].any(dim=1).sum().item()
            correct5 += correct[:, :5].any(dim=1).sum().item()
            total += y.size(0)

    end_time = time.time()
    avg_time_per_sample = (end_time - start_time) / total

    return {
        "Top-1 Accuracy": correct1 / total,
        "Top-3 Accuracy": correct3 / total,
        "Top-5 Accuracy": correct5 / total,
        "Avg Inference Time (s/sample)": avg_time_per_sample
    }
```

```python
results = evaluate(model, dataloader)

print("\n📊 Evaluation Results:")
for metric, value in results.items():
    print(f"{metric}: {value:.4f}")
```

```python
import math

class PositionalEncoding(nn.Module):
    def __init__(self, embed_dim, max_len=5000):
        super().__init__()
        pe = torch.zeros(max_len, embed_dim)
        position = torch.arange(0, max_len).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2) * -(math.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)  # shape: [1, max_len, embed_dim]
        self.register_buffer('pe', pe)

    def forward(self, x):
        # x: [batch_size, seq_len, embed_dim]
        x = x + self.pe[:, :x.size(1)].to(x.device)
        return x
```

```python
class TransformerModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, seq_len):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.pos_encoder = PositionalEncoding(embed_dim)
        encoder_layer = nn.TransformerEncoderLayer(
            d_model=embed_dim,
            nhead=4,
            dim_feedforward=512,
            dropout=0.1,
            activation="relu"
        )
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers=2)
        self.output = nn.Linear(embed_dim, vocab_size)

    def forward(self, x):
        # x: [batch_size, seq_len]
        x = self.embedding(x)                     # [B, S, E]
        x = self.pos_encoder(x)                   # [B, S, E]
        x = x.permute(1, 0, 2)                    # [S, B, E] for transformer
        x = self.transformer_encoder(x)           # [S, B, E]
        x = x[-1, :, :]                           # take last token's output: [B, E]
        return self.output(x)                     # [B, vocab_size]
```

```python
# Instantiate model
transformer = TransformerModel(VOCAB_SIZE, EMBED_DIM, SEQ_LEN).to(DEVICE)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(transformer.parameters(), lr=0.0001)  # smaller lr for transformers
```

```python
train(transformer, dataloader, optimizer, criterion, epochs=3)
```

```python
results = evaluate(transformer, dataloader)

print("\n📊 Transformer Evaluation Results:")
for metric, value in results.items():
    print(f"{metric}: {value:.4f}")
```

```python
def train(model, dataloader, optimizer, criterion, epochs=3):
    model.train()
    loss_history = []

    for epoch in range(epochs):
        total_loss = 0
        loop = tqdm(dataloader, desc=f"Epoch {epoch+1}")
        for X, y in loop:
            X, y = X.to(DEVICE), y.to(DEVICE)
            optimizer.zero_grad()
            outputs = model(X)
            loss = criterion(outputs, y)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            loop.set_postfix(batch_loss=loss.item())

        avg_loss = total_loss / len(dataloader)
        loss_history.append(avg_loss)
        print(f"[Epoch {epoch+1}] Avg Loss: {avg_loss:.4f}")

    return loss_history
```

```python
ffn_loss = train(model, dataloader, optimizer, criterion, epochs=3)

transformer = TransformerModel(VOCAB_SIZE, EMBED_DIM, SEQ_LEN).to(DEVICE)
optimizer = optim.Adam(transformer.parameters(), lr=0.0001)
transformer_loss = train(transformer, dataloader, optimizer, criterion, epochs=3)
```
